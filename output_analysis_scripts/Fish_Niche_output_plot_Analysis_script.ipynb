```python
import FishNiche_output_analysis_scripts2 as FN
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import date
import os
```

```python
lakes = FN.FishNiche_generate_timeseries("../test.csv",2,2)
```

```python
#set start and stop dates:
date_start = 2001
date_end = 2010
#set outputfile:
output= "..\\output"

timeaxis = pd.date_range ( "%s-01-01" % date_start, "%s-12-31" % date_end, freq="D" )
nlakes, colums = lakes.shape
i = 211
texte = "analyses_1"
plt.figure(nlakes+1)
for lakenum in np.arange ( 1, nlakes + 1 ):
    plt.subplot ( i )
    ts = pd.Series (lakes.loc[lakenum, "Volume with T < 15 C"], timeaxis )
    fig = ts.plot ( label="Volume with T < 15 C" )
    fig.set_xlabel ( "%s" % lakes.loc[lakenum, "Name"] )

    ts = pd.Series (lakes.loc[lakenum, "Volume with O2 > 3000"], timeaxis )
    fig = ts.plot ( label="Volume with O2 > 3000" )

    if lakenum == nlakes:
        plt.legend ( bbox_to_anchor=(1.02, 2.8
                                    ), loc="upper right",
                        ncol=2 )

    plt.tight_layout ()
    i += 1

plt.savefig ( "%s\\%s_graph.pdf" % (output, texte) )
plt.show()
```
